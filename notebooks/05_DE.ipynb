```python
import os
import sys 
import logging

# Caminho para a raiz do seu projeto
project_root = os.path.abspath("..")  # Sobe um nível de pasta a partir de 'notebooks'

# Adiciona o caminho ao sys.path
if project_root not in sys.path:
    sys.path.insert(0, project_root)

# Função para listar todos os arquivos e pastas
def listar_arquivos_pastas(project_root):
    for raiz, diretorios, arquivos in os.walk(project_root):
        print(f'Pasta: {raiz}')
        for nome in arquivos:
            print(f'  Arquivo: {nome}')
        for nome in diretorios:
            print(f'  Subpasta: {nome}')

# Chamando a função
listar_arquivos_pastas(project_root)
```
